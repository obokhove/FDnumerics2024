Consider the Poisson system
$$
-\nabla ^2 u = f \text{ on } (x,y) \in [0,1]^2 \tag{1-a}
$$
$$
f(x,y) = 2\pi^2\sin{(\pi x)}\cos{(\pi y)} \tag{1-b}
$$
$$
u(0,y) = u(1,y) = 0 \tag{1-c}
$$
$$
\frac{\partial u(x,0)}{\partial y} = \frac{\partial u(x,1)}{\partial y} = 0 \tag{1-d}
$$
with variable or unknown $u(x,y)$, given function $f(x,y)$, and with Dirichlet and Neumann boundary conditions. The exact solution is
$$
u_e(x,y) = \sin(\pi x)\cos(\pi y)
$$

First, we let $u$ satisfy the following minimisation problem
$$
\underset{u \in \Omega}{\min} \space F[u] \space ; F[u] = \int _0^1 \int _0^1 \{\frac{1}{2}|\nabla u|^2 - uf\}dxdy \tag{2}
$$
consider $u = \tilde{u} + \epsilon \delta u$, where $\epsilon$ is a variable parameter, $\tilde{u}(x,y)$ is a smooth solution of the minimisation problem (2) above, with a fixed variation $\delta u(x,y)$. If the parameter $\epsilon$ is 0, $\tilde{u}$ is a solution to the minimisation problem. A necessary condition for the existence of a minimum solution is given by
$$
\frac{dF[u]}{d\epsilon} = 0 \tag{2}
$$
therefore by substituting $u = \tilde{u} + \epsilon \delta u$ into the minimisation problem and taking the derivative w.r.t $\epsilon$
$$
\frac{d}{d\epsilon}\left[\int _0^1 \int _0^1 \{\frac{1}{2}|\nabla (\tilde{u} + \epsilon \delta u)|^2 - (\tilde{u} + \epsilon \delta u)f\}dxdy \right] = 0 \tag{3}
$$
$$
\int _0^1 \int _0^1 \{ \frac{d}{d\epsilon} \left[\frac{1}{2} \left(\frac{d(\tilde{u} + \epsilon \delta u)}{dx}\right)^2 +  \frac{1}{2}\left(\frac{d((\tilde{u} + \epsilon \delta u))}{dy} \right)\right] - \frac{d}{d\epsilon}\left[(\tilde{u} + \epsilon \delta u)f\right]\}dxdy = 0 \tag{4}
$$
$$
\int _0^1 \int _0^1 \{\nabla(\tilde{u} + \epsilon \delta u)\cdot \nabla(\delta u) - f\delta u \} dxdy = 0 \tag{5}
$$
Recalling that if $\epsilon$ is 0 then $\tilde{u}$ is the minimal solution, (5) reduces to
$$
\int _0^1 \int _0^1 \{\nabla\tilde{u}\cdot \nabla(\delta u) - f\delta u \} dxdy = 0 \tag{6}
$$
If we perform integration by parts w.r.t $x$ and $y$, we can substitute the boundary conditions to get a reduced expression as follows
$$
\int _0^1 \left\{ \left[\frac{\partial \tilde{u}}{\partial x} \delta u \right]_{x=0}^{x=1} - \int _0^1 \{ \frac{\partial^2 \tilde{u}}{\partial x^2} \delta u\} dx\right\} dy +  \int _0^1 \left\{ \left[\frac{\partial \tilde{u}}{\partial y} \delta u \right]_{y=0}^{y=1} - \int _0^1 \{\frac{\partial^2 \tilde{u}}{\partial y^2} \delta u \}dy\right\} dx - \int _0^1 \int _0^1 \{\delta u f \}dxdy = 0 \tag{7}
$$
$$
\int _0^1 \int _0^1 \{\delta u(-\nabla^2\tilde{u} - f) \}dxdy = 0 \tag{8}
$$
or in another form
$$
\int _0^1 \int _0^1 \{\nabla u \cdot \nabla \delta u - \delta u f \}dxdy = 0 \tag{9}
$$
This is the derivation of the variation of the Ritz Galerkin principle.

We can also derive the weak formulation of system (1-a) by multiplying the strong form of the differential equation by an arbitrary test function $w(x,y)$, and then integrating over the doman $\Omega$
$$
- \int _0^1 \int _0^1 \{w \nabla^2 u \}dxdy = \int _0^1 \int _0^1 \{w f\}dxdy \tag{10}
$$
$$
- \int _0^1 \int _0^1 \{\nabla \cdot(w\nabla u) \}dxdy +  \int _0^1 \int _0^1 \{\nabla w \cdot \nabla u \}dxdy = \int _0^1 \int _0^1 \{w f\}dxdy \tag{11}
$$
By applying the divergence theorem 
$\iint_V (\nabla \cdot \mathbf{F}) \, dV = \oint\oint_{S}( \mathbf{F} \cdot \mathbf{n}) dS$, we get
$$
\int _0^1 \int _0^1 \{\nabla w \cdot \nabla u \}dxdy - \oint \{w \nabla u \cdot \hat{n} \}d \Gamma = \int _0^1 \int _0^1 \{w f\}dxdy \tag{12}
$$

In order to satisy the boundary conditons, the boundary integral $\oint \{w \nabla u \cdot \hat{n} \}d \Gamma$ needs to be evaluated.

- On $\Gamma _1 : \nabla u \cdot \hat{n} = -\frac{\partial u}{\partial x}, u(0,y) = 0 \implies \frac{\partial u}{\partial x} = 0$
- On $\Gamma _2 : \nabla u \cdot \hat{n} = \frac{\partial u}{\partial x} = 0$
- On $\Gamma _3 : \nabla u \cdot \hat{n} = -\frac{\partial u}{\partial y}, \frac{\partial u(x,0)}{\partial y} = 0 \implies \frac{\partial u}{\partial y} = 0$
- On $\Gamma _4 : \nabla u \cdot \hat{n} = \frac{\partial u}{\partial y}, \frac{\partial u(x,1)}{\partial y} = 0 \implies \frac{\partial u}{\partial y} = 0$
\

Therefore, the boundary integral is reduced to 0, and the weak form is given by
$$
\int _0^1 \int _0^1 \{\nabla u \cdot \nabla w  - w f\}dxdy = 0 \tag{13}
$$
We can see that (9) given by the Ritz-Galerkin principle and (13) given by the weak formulation are identical, where $\delta u (x,y) = w(x,y)$.

The above formulation gives the continuous form of the Ritz Galerkin principle and weak form for the Poisson system (1). In order to implement this into a Finite Element Method solver, a discrete version is required. We introduce a FEM expasion in the form
$$
u(x,y) \approx u_h(x,y)
$$
where $u_x(x,y)$ is an approximate solution given by a linear combination of basis/shape functions
$$
u_h(x,y) = \sum _{j=1}^N \hat{u}_j \phi _j(x,y) \tag{14}
$$
where $\hat{u}_j$ is a coefficient and $\phi _j(x,y)$ is a proposed basis function, summed over N nodes. We can also susbstitute for the variation 
$$
\partial u \approx \sum_{j=1}^N \phi _j(x,y) \tag{15}
$$
We can substitute these into the equations for the continuous Ritz-Galerkin principle (9)
$$
F[u_h] = \int _0^1 \int _0^1\{ \frac{1}{2}|\sum_{j=1}^N \hat{u}_j\nabla \phi _j|^2 - \sum _{j=1}^N \phi _j f\} dxdy = 0 \tag{16}
$$
and the weak form (13)
$$
\int _0^1 \int _0^1 \{(\sum _{j=1}^N \nabla \phi _j)(\sum _{j=1}^N \hat{u}_j \nabla \phi) - \sum _{j=1}^N \phi _j f\}dxdy \tag{17}
$$
To show that the variation of one leads to the other, we can apply a  variation to the Ritz-Galerkin principle where $\partial u \approx \sum _{j=1}^N \phi _j(x,y)$
$$
F\left[u_h + \epsilon \sum _{j=1}^N \phi _j\right] = \int _0^1 \int _0^1 \{\frac{\partial}{\partial t} \left(\frac{1}{2} \nabla|\sum _{j=1}^N \hat{u}_j\phi _j  + \epsilon \sum _{j=1}^N \phi _j|^2 \right) - \sum_{j=1}^N \phi_j f \}dxdy \tag{18}
$$
Expanding this bracket, and assuming that the $\epsilon$ term is disregarded, we are left with
$$
\frac{d}{dt}F\left[u_h + \epsilon \sum _{j=1}^N \phi _j\right] = \int _0^1 \int _0^1 \{\nabla \left(\sum _{j=1}^N \hat{u}_j \phi _j \sum _{j=1}^N \phi _j \right)-\sum _{j=1}^N \phi _j f\}dxdy \tag{19}
$$
which equates to
$$
\frac{d}{dt}F\left[u_h + \epsilon \sum _{j=1}^N \phi _j\right] = \int _0^1 \int _0^1 \{  \left(\sum _{j=1}^N \nabla \phi _j \right) \left(\sum _{j=1}^N \hat{u}_j \nabla \phi _j  \right) - \sum _{j=1}^N \phi _j f\}dxdy \tag{20}
$$
(20) is the same as the discrete version of the weak form (17).

The system is solved in Firedrake using both the Ritz-Galerkin principle and the weak form method.

```python
#
# Solution by Firedrake FEM-CG of a Poisson equatiom
#
from firedrake import *

nx = ny = 256 # Try various mesh resolutions, starting coarse, say 16x16 etc.

mesh = UnitSquareMesh(nx,ny,quadrilateral=True)
# Quadrilateral regular mesh made: https://www.firedrakeproject.org/firedrake.html#firedrake.utility_meshes.UnitSquareMesh
# Alternatively use gmsh: 

V = FunctionSpace(mesh, 'CG', 3) # Piecewise linear continuous Galerkin function space or polynomials
# See: https://www.firedrakeproject.org/variational-problems.html

#
# Method 1: construct the weak form manually by multiplying and manipulating the Poisson equation and solve the linear system
#
u = TrialFunction(V) # The unknown or variable u(x,y)
v = TestFunction(V)  # The testfunction of u, which may be better called delu or deltau

x, y = SpatialCoordinate(mesh) # Mesh coordinates

f = Function(V).interpolate(2*pi**2*sin(pi*x)*cos(pi*y)) # The given function f(x,y)

a = (inner(grad(u),grad(v)))*dx # Step 2/3: The weak form first term
L = (f*v)*dx # Step 2/3: The weak form second term; dx is the infinitesimal piece in the damain here: dx*dy=dA with area A.

u_1 = Function(V, name='u_1') # Name of solution for first method

bc_x0 = DirichletBC(V, Constant(0), 1) # Dirichlet boundary conditions imposed 
bc_x1 = DirichletBC(V, Constant(0), 2) # Dirichlet boundary conditions imposed 
# See: https://www.firedrakeproject.org/firedrake.html#firedrake.bcs.DirichletBC
# The homogeneous Neumann boundary conditions are "automatically" included, i.e. do not need anything explicit

solve(a == L, u_1, solver_parameters={'ksp_type': 'cg', 'pc_type': 'none'}, bcs=[bc_x0,bc_x1]) # Step 4: the solution assigned to u1

#
# Method 2: generate the weak form via "derivative()" of the Ritz-Galerkin integral or variational principle and solve the nonlinear system
#
u_2 = Function(V, name='u_2') # Name of solution for first method

Ju = (0.5*inner(grad(u_2),grad(u_2)) - u_2*f)*dx # f->ULF? Step 2

F = derivative(Ju, u_2, du=v) # Step 2/3: The weak form generated

solve(F == 0, u_2, bcs=[bc_x0, bc_x1]) # Step 4: the solution assigned to u2

# Post-processing: Use Paraview to visualize
# First, assign the differences to new Function objects
f.interpolate(sin(pi*x)*cos(pi*y))
u_1_minus_f = Function(V)
u_2_minus_f = Function(V)

u_1_minus_f.assign(u_1 - f)
u_2_minus_f.assign(u_2 - f)

# 
# Post-processing: Use Paraview to visualise
# See https://www.firedrakeproject.org/visualisation.html#creating-output-files
outfile = VTKFile('output.pvd')
outfile.write(u_1_minus_f, u_2_minus_f)


L2_1 = sqrt(assemble(dot(u_1 - f, u_1 - f) * dx)) # L2 error solution u1
L2_2 = sqrt(assemble(dot(u_2 - f, u_2 - f) * dx)) # L2 error solution u2
L2 = sqrt(assemble(dot(u_2 - u_1, u_2 - u_1) * dx)) # L2 error difference
print(f'Mesh resolution: Î”x = {1/nx}')
print(f'L2 error: Method1 = {L2_1}, Method2 = {L2_2}')
print(f'L2 norm between the two results: {L2}')
```

This produces the following contour plot, which can be viewed in Paraview. For a grid size of 32x32, the Ritz-Galerkin produces an error $|u_h - u_e|$ of $0.00040075733647807485$, and the weak form produces an error of $0.00040075733647211825$, with an error between the two methods of $6.008347830603989e-15$, which is essentially machine precision since we have demonstrated that the two methods are equivalent.
