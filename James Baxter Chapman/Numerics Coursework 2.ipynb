Since we have the solutions to these Riemann problems as calculated in Question (2), we can piece them together over the whole domain. This means that rather than a continuous solution, we will have discrete approximate solutions at each cell defined by the exact solution at the cell centre given by
$$
U_j^{n+1} = \frac{1}{\Delta x}\int _{x_{j-\frac{1}{2}}}^{x_{j+\frac{1}{2}}} \tilde{u}^n(x, t^{n+1})dx
$$
where $\tilde{u}^n(x,t)$ is the continuous solution for $u$ at position $x$ and time $t$. \
This can be simplified by using the integral/weak form of the conservation law. Since $\tilde{u}^n$ is assumed to be an exact weak solution, we know that for an upwind scheme
$$
\int _{x_{j-\frac{1}{2}}}^{x_{j+\frac{1}{2}}} \tilde{u}^n(x, t^{n+1})dx = \int _{x_{j-\frac{1}{2}}}^{x_{j+\frac{1}{2}}} \tilde{u}^n(x, t^{n})dx + \int _{t^n}^{t^{n+1}}f(\tilde{u}^n(x_{j-\frac{1}{2}},t))dt - \int _{t^n}^{t^{n+1}}f(\tilde{u}^n(x_{j+\frac{1}{2}},t))dt
$$
Dividing by $\Delta x$ and using the discrete form noting that $\tilde{u}^n(x,t^n) \equiv U_j^n$ over the cell volume, the Godunov scheme is given by
$$
\bar{U}_j^{n+1} = \bar{U}_j^n - \frac{\Delta t^n}{\Delta x_j}\left(\bar{F}_{j+1}(\bar{U}_j^n , \bar{U}_{j+1}^n) - \bar{F}_j(\bar{U}_{j-1}^n, \bar{U},j^n) \right) \tag{11}
$$
in it's discrete form, where j is indexing for volumes. The flux function $F$ is given by
$$
F(U_j^n,U_{j+1}^n) = \frac{1}{\Delta t^{n+1}}\int _{t^n}^{t^{n+1}}f(\tilde{u}^n(x_{j+\frac{1}{2}},t))dt \tag{12}
$$
This is simply equal to the value for the zone in between $-c_0 t \le x \le c_0 t$, depending on the function $f(u)$.

$$
Which is essentialy equivalent to the Courant number, which was discussed in Coursework 1.
