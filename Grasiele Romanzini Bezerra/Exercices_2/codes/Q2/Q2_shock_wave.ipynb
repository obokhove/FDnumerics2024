```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
h_plot = 1
v_plot = 1
colours = sns.color_palette()
colour_index = [1,4,9]

fig, ax = plt.subplots(v_plot, h_plot, figsize=(10, 3))
## particle_path

plt.vlines(x = 0, ymin = 0, ymax = 0.5, color = colours[1])
plt.hlines(y = 0, xmin = -10, xmax = 0, color = colours[1])
plt.hlines(y = 0.5, xmin = 0, xmax = 10, color = colours[1])

plt.vlines(x = 2, ymin = 0, ymax = 0.5, color = colours[4], ls = 'dotted')
plt.hlines(y = 0, xmin = -10, xmax = 2, color = colours[4], ls = 'dotted')
plt.hlines(y = 0.5, xmin = 2, xmax = 5, color = colours[4],  ls = 'dotted')

plt.vlines(x = -2, ymin = 0, ymax = 0.5, color = colours[9],  ls = 'dotted')
plt.hlines(y = 0, xmin = -10, xmax = -2, color = colours[9],  ls = 'dotted')
plt.hlines(y = 0.5, xmin = -2, xmax = 10, color = colours[9],  ls = 'dotted')

sns.despine(top=True, right=True, left=True, bottom=True)

ax.set_xticks([-2,0,2])
ax.set_yticks([0,0.5])
ax.set_xticklabels([r'$-c_0t$', 0, r'$c_0t$'])
ax.set_yticklabels([r'$r_{1l}$, $r_{2l}$',r'$r_{1r}$, $r_{2r}$'])
ax.text(0.1,0.2, r"$r_1$ and $r_2$ at $t = 0$ have the discontinuity at x = 0", color = colours[1])
ax.text(2.1,0.1, r"$r_1$ have discontinuity at $x = c_0t$ ", color = colours[4])
ax.text(2.1,0.05, r"The shock wave propagates to the right", color = colours[4])
ax.text(-8.8,0.1, r"$r_2$ have discontinuity at $x = -c_0t$", color = colours[9])
ax.text(-8.8,0.05, r"The shock wave propagates to the left", color = colours[9])


ax.annotate('', xy=(0.0, 0.6), xytext=(0, 0),
            arrowprops=dict(facecolor='black', arrowstyle='->'),
            xycoords=('axes fraction', 'data'), textcoords='axes fraction')

ax.annotate('', xy=(10, 0), xytext=(0, 0),
            arrowprops=dict(facecolor='black', arrowstyle='->'),
            xycoords=('data', 'axes fraction'), textcoords='axes fraction')
ax.set_xlabel("x")
plt.savefig('images/Question2_shock_wave.png', dpi = 500, bbox_inches = 'tight')
```
